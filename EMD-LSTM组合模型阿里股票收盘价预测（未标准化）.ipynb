```python
import pandas as pd
import numpy as np
from PyEMD.EMD import EMD
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 读取数据
def load_data(filename, sheet_name):
    df = pd.read_excel(filename, sheet_name=sheet_name, parse_dates=True, index_col=0)
    return df['Close'], df.index  # 返回收盘价和索引

# 进行 EMD 分解
def perform_emd(data):
    emd = EMD()  # 创建 EMD 实例
    imfs = emd.emd(data.values)  # 使用 emd 方法，并确保传入数值数组
    return imfs

# LSTM模型
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out

# 数据预处理
def preprocess_data(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length])
    return np.array(X), np.array(y)

# 训练函数
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs.squeeze(), targets)
            loss.backward()
            optimizer.step()
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# 评估函数
def evaluate_model(model, data_loader):
    model.eval()
    predictions = []
    targets = []
    with torch.no_grad():
        for inputs, t in data_loader:
            outputs = model(inputs)
            predictions.append(outputs.numpy())
            targets.append(t.numpy())
    predictions = np.concatenate(predictions)
    targets = np.concatenate(targets)
    return predictions, targets

# 可视化结果
def plot_results(targets_train, predictions_train, targets_test, predictions_test, imf_index, index):
    plt.figure(figsize=(12, 6))
    plt.plot(index[:len(targets_train)], targets_train, label='Actual Train', linestyle='--', color='blue')
    plt.plot(index[:len(predictions_train)], predictions_train, label='Predicted Train', color='red')
    plt.plot(index[len(targets_train):len(targets_train) + len(targets_test)], targets_test, label='Actual Test', linestyle='--', color='green')
    plt.plot(index[len(predictions_train):len(predictions_train) + len(predictions_test)], predictions_test, label='Predicted Test', color='orange')
    plt.title(f'IMF {imf_index + 1} Prediction')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.xticks(rotation=45)  # 使x坐标标记旋转以便更清晰
    plt.legend()
    plt.tight_layout()  # 自动调整 subplot 参数以给标签留出空间
    plt.show()

if __name__ == "__main__":
    # 加载数据
    filename = 'stocks_data_alphavantage.xlsx'  # 数据文件路径
    sheet_name = 'Alibaba'  # 工作表名称
    close_prices, index = load_data(filename, sheet_name)

    # 进行 EMD 分解
    imfs = perform_emd(close_prices)

    # 合并IMF7和IMF8
    # if len(imfs) >= 8:  # 确保有足够的IMF
    #     imf1 = imfs[0]  # IMF 1 (索引0)
    #     imf7 = imfs[6]  # IMF 7 (索引6)
    #     imf8 = imfs[7]  # IMF 8 (索引7)
    #     new_imf = imf1 + imf7 + imf8
    #     imfs[0] = new_imf  # 用新IMF替换IMF0
    # imfs = np.delete(imfs, [6, 7], axis=0)  # 删除原IMF7,IMF8

    # 更新IMF数量
    num_imfs = len(imfs)

    # 可视化原始数据和IMF
    plt.figure(figsize=(12, 4 * num_imfs))  # 设置图形高度为4 * num_imfs
    
    # 在同一个图形窗口中绘制所有IMF
    for i in range(num_imfs):
        plt.subplot(num_imfs, 1, i + 1)  # 更新为num_imfs行1列
        plt.plot(index[:len(imfs[i])], imfs[i], label=f'IMF {i + 1}')
        plt.title(f'IMF {i + 1}')
        plt.xlabel('Date')
        plt.ylabel('Value')
        plt.xticks(rotation=45)
        plt.legend()
        plt.tight_layout()

    # 显示图形
    plt.show()

    # LSTM建模并训练每个IMF分量
    sequence_length = 30
    input_size = 1
    hidden_size = 50
    num_layers = 2
    output_size = 1
    num_epochs = 50
    batch_size = 16

    all_predictions_train = []
    all_predictions_test = []
    all_targets_train = []
    all_targets_test = []

    for i in range(num_imfs):
        imf = imfs[i]

        # 处理IMF数据
        X, y = preprocess_data(imf, sequence_length)
        X = X.reshape(X.shape[0], X.shape[1], input_size)

        # 划分训练集和测试集
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, shuffle=False)

        train_dataset = TensorDataset(torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.float32))
        test_dataset = TensorDataset(torch.tensor(X_test, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32))

        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

        model = LSTMModel(input_size, hidden_size, num_layers, output_size)
        criterion = nn.MSELoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        # 训练模型
        print(f'Training model for IMF {i + 1}...')
        train_model(model, train_loader, criterion, optimizer, num_epochs)

        # 评估模型
        predictions_train, targets_train = evaluate_model(model, train_loader)
        predictions_test, targets_test = evaluate_model(model, test_loader)

        # 打印评估报告
        mse_train = mean_squared_error(targets_train, predictions_train)
        mae_train = mean_absolute_error(targets_train, predictions_train)
        r2_train = r2_score(targets_train, predictions_train)
        print(f'IMF {i + 1} Train - MSE: {mse_train:.4f}, MAE: {mae_train:.4f}, R²: {r2_train:.4f}')

        mse_test = mean_squared_error(targets_test, predictions_test)
        mae_test = mean_absolute_error(targets_test, predictions_test)
        r2_test = r2_score(targets_test, predictions_test)
        print(f'IMF {i + 1} Test - MSE: {mse_test:.4f}, MAE: {mae_test:.4f}, R²: {r2_test:.4f}')

        # 可视化预测结果
        plot_results(targets_train, predictions_train, targets_test, predictions_test, i, index)

        # 保存预测结果
        all_predictions_train.append(predictions_train)
        all_predictions_test.append(predictions_test)
        
        all_targets_train.append(targets_train)
        all_targets_test.append(targets_test)

    # 合并所有IMF的预测结果
    combined_predictions_train = np.sum(all_predictions_train, axis=0)
    combined_predictions_test = np.sum(all_predictions_test, axis=0)
    
    combined_targets_train = np.sum(all_targets_train, axis=0)
    combined_targets_test = np.sum(all_targets_test, axis=0)

    # 评估合并后的预测结果
    mse_train_combined = mean_squared_error(combined_targets_train, combined_predictions_train)
    mae_train_combined = mean_absolute_error(combined_targets_train, combined_predictions_train)
    r2_train_combined = r2_score(combined_targets_train, combined_predictions_train)
    print(f'Combined IMF Train - MSE: {mse_train_combined:.4f}, MAE: {mae_train_combined:.4f}, R²: {r2_train_combined:.4f}')

    mse_test_combined = mean_squared_error(combined_targets_test, combined_predictions_test)
    mae_test_combined = mean_absolute_error(combined_targets_test, combined_predictions_test)
    r2_test_combined = r2_score(combined_targets_test, combined_predictions_test)
    print(f'Combined IMF Test - MSE: {mse_test_combined:.4f}, MAE: {mae_test_combined:.4f}, R²: {r2_test_combined:.4f}')

    # 可视化合并后的预测结果
    plot_results(combined_targets_train, combined_predictions_train, combined_targets_test, combined_predictions_test, -1, index)
```
